```python
%%manim -qm -v WARNING SolveQuadratic

class SolveQuadratic(Scene):
    def construct(self):
        # Introdução: Mostrar a equação inicial
        equation = MathTex("x^2 - 6x + 8 = 0")
        self.play(Write(equation))
        self.wait(1)

        # Passo 1: Fatoração
        step1 = MathTex("x^2 - 6x + 8", "=", "(x - 2)(x - 4)").next_to(equation, DOWN)
        self.play(TransformMatchingTex(equation.copy(), step1))
        self.wait(1)

        # Passo 2: Zerar os fatores
        step2 = MathTex("(x - 2)(x - 4)", "=", "0").next_to(step1, DOWN)
        self.play(TransformMatchingTex(step1.copy(), step2))
        self.wait(1)

        # Passo 3: Solução dos fatores
        step3 = MathTex("x - 2 = 0 \\quad \\text{ou} \\quad x - 4 = 0").next_to(step2, DOWN)
        self.play(TransformMatchingTex(step2.copy(), step3))
        self.wait(1)

        # Passo 4: Solução final
        solution = MathTex("x = 2 \\quad \\text{ou} \\quad x = 4").next_to(step3, DOWN)
        self.play(TransformMatchingTex(step3.copy(), solution))
        self.wait(2)

        # Destacar o resultado final
        final_box = SurroundingRectangle(solution, color=YELLOW, buff=0.2)
        self.play(Create(final_box))
        self.wait(2)

        # Encerrar
        self.play(FadeOut(equation, step1, step2, step3, solution, final_box))
```

```python
%%manim -qm -v WARNING MultipleSvgAnimation

class MultipleSvgAnimation(Scene):
    def construct(self):
        # Lista com os caminhos para os arquivos SVG
        comom_path = "/content/SVG/Logo"
        svgtype = "white.svg"
        eixoKeys = ["PC", "PG", "IB", "FI", "BIO", "QI"]  # Substitua pelos seus arquivos

        # Ajuste de tamanho e deslocamento
        svg_height = 2  # Altura das imagens
        start_pos = UP * 5  # Posição inicial (fora da tela acima)
        end_pos = DOWN * 5  # Posição final (fora da tela abaixo)

        for eixo in eixoKeys:
            svg_image = SVGMobject(comom_path + eixo + svgtype)

            svg_image.set_height(svg_height)  # Ajusta o tamanho da imagem
            svg_image.move_to(start_pos)  # Posiciona fora da tela no topo

            # Animação para mover do topo para baixo
            self.play(svg_image.animate.move_to(ORIGIN))  # Move para o centro
            self.wait(0.5)  # Pausa breve no centro
            self.play(svg_image.animate.move_to(end_pos))  # Move para fora da tela abaixo
```

```python
%%manim -qm -v WARNING SlotMachineEffect

class SlotMachineEffect(Scene):
    def construct(self):
        # Defina as imagens SVG (substitua pelos caminhos das suas imagens SVG)
        svg_files = ["/content/SVG/LogoFIwhite.svg",
                     "/content/SVG/LogoPGwhite.svg",
                     "/content/SVG/LogoPCwhite.svg",
                     "/content/SVG/LogoQIwhite.svg",
                     "/content/SVG/LogoIBwhite.svg",
                     "/content/SVG/LogoBIOwhite.svg"]

        svg_files = svg_files * 10

        # Criar uma lista de objetos SVG
        svgs = [SVGMobject(file).set_height(2) for file in svg_files]

        # Empilhar as imagens com espaçamento entre elas
        for i, svg in enumerate(svgs):
            svg.move_to(UP * (i * 3))  # 3 unidades de distância entre cada imagem

        # Agrupar as imagens para animá-las em conjunto
        svg_group = VGroup(*svgs)

        # Adicionar o grupo de imagens à cena
        self.add(svg_group)

        # Animação: deslizar as imagens para cima, criando o efeito de slot machine
        self.play(svg_group.animate.shift(DOWN * 15), run_time=2, rate_func=linear)

        # Voltar ao topo e repetir (simulando o efeito de "rolagem")
        self.play(svg_group.animate.shift(UP * 15), run_time=0.1)

        # Animação: deslocamento contínuo com repetição (ainda mais como o efeito de slot)
        self.play(svg_group.animate.shift(DOWN * 15), run_time=2, rate_func=linear)

        # Parar em uma posição aleatória
        self.wait(1)
```

```python
%%manim -qm -v WARNING SlotMachineEffect1

class SlotMachineEffect1(Scene):
    def construct(self):
        # Defina as imagens SVG (substitua pelos caminhos das suas imagens SVG)
        svg_files = ["/content/SVG/LogoNivelamentoWhite.svg",
                    "/content/SVG/LogoFIwhite.svg",
                     "/content/SVG/LogoPGwhite.svg",
                     "/content/SVG/LogoPCwhite.svg",
                     "/content/SVG/LogoQIwhite.svg",
                     "/content/SVG/LogoIBwhite.svg",
                     "/content/SVG/LogoBIOwhite.svg"]

        svg_files = svg_files * 10

        # Criar uma lista de objetos SVG
        svgs = [SVGMobject(file).set_height(2) for file in svg_files]

        # Empilhar as imagens com espaçamento entre elas
        for i, svg in enumerate(svgs):
            svg.move_to(UP * (i * 3))  # 3 unidades de distância entre cada imagem

        # Agrupar as imagens para animá-las em conjunto
        svg_group = VGroup(*svgs)

        # Adicionar o grupo de imagens à cena
        self.add(svg_group)

        # Função de aceleração e desaceleração personalizada
        def acceleration_deceleration(t):
            if t < 0.5:
                return 2 * t ** 2  # Acelera até a metade
            else:
                return -2 * (t - 1) ** 2 + 1  # Desacelera após a metade

        # Duração total da animação cíclica
        total_run_time = 10  # 10 segundos de animação

        # Definir número de ciclos para a animação
        num_cycles = 3  # Número de ciclos a serem realizados durante os 10 segundos

        # Animação cíclica: repetir a rotação da slot machine durante o tempo desejado
        self.play(
            svg_group.animate.shift(DOWN * 15),
            run_time=total_run_time,  # A animação vai durar 10 segundos
            rate_func=acceleration_deceleration,  # Função personalizada para aceleração/desaceleração
            repeat=True,  # Repetir a animação de forma cíclica
            cycle=total_run_time / (total_run_time / num_cycles)  # Definir a quantidade de ciclos a serem feitos
        )

        # Pausar no final da animação para ver o resultado
        self.wait(1)
```

```python
%%manim -qm -v WARNING LogoAnimation
import random

from manim import *
import random

class LogoAnimation(Scene):
    def construct(self):
        # Lista de caminhos para as logos SVG
        svg_files = [
            "/content/SVG/LogoNivelamentoWhite.svg",
            "/content/SVG/LogoFIwhite.svg",
            "/content/SVG/LogoPGwhite.svg",
            "/content/SVG/LogoPCwhite.svg",
            "/content/SVG/LogoQIwhite.svg",
            "/content/SVG/LogoIBwhite.svg",
            "/content/SVG/LogoBIOwhite.svg"
        ]

        # Carregar as logos como objetos SVG
        logos = [SVGMobject(file) for file in svg_files]

        # Definir a posição inicial da animação
        logo_positions = [UP * 3, UP * 2, UP, DOWN, DOWN * 2, DOWN * 3]

        # Estabelecer a animação de carrossel infinito
        def move_logo(logo, start_position, end_position, speed):
            logo.move_to(start_position)
            self.play(logo.animate.move_to(end_position).rate_func(lambda t: smooth(1 - (t**2))), run_time=speed)

        # Mostrar a logo inicial
        initial_logo = logos[0]
        initial_logo.move_to(ORIGIN)
        self.play(FadeIn(initial_logo))

        # Aumento da velocidade das logos conforme o tempo
        speed = 2  # Velocidade inicial
        logo_index = 0
        cycles = 10  # Número de ciclos do carrossel

        for cycle in range(cycles):
            # Escolher logo aleatória
            random_logo = random.choice(logos)

            # Mover logo de cima para baixo
            move_logo(random_logo, UP * 3, DOWN * 3, speed)

            # Aumentar a velocidade a cada ciclo
            speed = max(0.5, speed - 0.1)  # A velocidade diminui gradualmente

        # Quando parar na logo específica (LogoPCwhite.svg)
        target_logo = logos[3]  # Logo "/content/SVG/LogoPCwhite.svg"

        # Reduzir a velocidade até aparecer a logo desejada
        slow_down_speed = 5
        move_logo(target_logo, UP * 3, DOWN * 3, slow_down_speed)

        # Fim da animação
        self.wait(1)

```
